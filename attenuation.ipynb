```python
import numpy as np
import matplotlib.pyplot as plt

# Initial intensity
I0 = 1.0  

# Material attenuation coefficients (per cm)
# Approx realistic  values for gamma energies ~0.51 MeV
mu = {
    "Air": 0.0001,
    "Water": 0.05,
    "Concrete": 0.12,
    "Lead": 1.25
}

# Thickness range (020 cm)
x = np.linspace(0, 20, 300)

plt.figure(figsize=(8, 6))

for material, mu_value in mu.items():
    I = I0 * np.exp(-mu_value * x)
    plt.plot(x, I, label=material, linewidth=2)

plt.title("Gamma Radiation Attenuation Through Different Materials", fontsize=14)
plt.xlabel("Thickness (cm)", fontsize=12)
plt.ylabel("Remaining Intensity (I/I0)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend()
plt.tight_layout()
plt.show()
```
